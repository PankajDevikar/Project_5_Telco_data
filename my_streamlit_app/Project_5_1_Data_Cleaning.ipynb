**Replace value with zero**

```python
def drop_high_missing_columns(data, threshold):
    missing_proportions = data.isnull().sum() / len(data)
    columns_to_drop = missing_proportions[missing_proportions >= threshold].index
    return data.drop(columns_to_drop, axis=1)
```

```python
df = drop_high_missing_columns(data, 0.7)
df.columns
```

```python
def remove_missing_values():
    columns_to_check = [
    'Start',
    'Start ms',
    'End',
    'End ms',
    'Dur. (ms)',
    'Avg Bearer TP DL (kbps)',
    'Avg Bearer TP UL (kbps)',
    'Activity Duration DL (ms)',
    'Activity Duration UL (ms)',
    'Dur. (ms).1',
    'Total UL (Bytes)',
    'Total DL (Bytes)'
    ]

    return df.dropna(subset=columns_to_check)

cleaned_df = remove_missing_values()
```

**Impute columns with the mean value**

```python
def impute_columns():
    columns_to_impute = [
        "Avg RTT DL (ms)",
        "Avg RTT UL (ms)",
        "TCP DL Retrans. Vol (Bytes)",
        "TCP UL Retrans. Vol (Bytes)",
        "DL TP < 50 Kbps (%)",
        "50 Kbps < DL TP < 250 Kbps (%)",
        "250 Kbps < DL TP < 1 Mbps (%)",
        "DL TP > 1 Mbps (%)",
        "UL TP < 10 Kbps (%)",
        "10 Kbps < UL TP < 50 Kbps (%)",
        "50 Kbps < UL TP < 300 Kbps (%)",
        "UL TP > 300 Kbps (%)",
        "HTTP DL (Bytes)",
        "HTTP UL (Bytes)",
        "Nb of sec with 125000B < Vol DL",
        "Nb of sec with 1250B < Vol UL < 6250B",
        "Nb of sec with 31250B < Vol DL < 125000B",
        "Nb of sec with 6250B < Vol DL < 31250B",
        "Nb of sec with Vol DL < 6250B",
        "Nb of sec with Vol UL < 1250B"
    ]

    imputer = SimpleImputer(strategy='mean')
    cleaned_df[columns_to_impute] = imputer.fit_transform(cleaned_df[columns_to_impute])
    return cleaned_df
imputed_df = impute_columns()
```

**Replace columns with mode value**

```python
def replace_with_mode():
    columns_to_replace = [
        'Bearer Id',
        'IMSI',
        'MSISDN/Number',
        'IMEI',
        'Last Location Name',
        'Handset Manufacturer',
        'Handset Type'
    ]
    imputer = SimpleImputer(strategy='most_frequent')
    cleaned_df[columns_to_replace] = imputer.fit_transform(cleaned_df[columns_to_replace])
    return cleaned_df

cleaned_df = replace_with_mode()
```

**Remove outliers**

```python
def handle_outliers(cleaned_df, columns):
    df = cleaned_df.copy()
    for col in columns:
        # Calculate outlier bounds (replace with your chosen method)
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Replace outliers with NaN (or any other strategy)
        df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound),
                           np.nan, df[col])
    return df
```

```python
columns = [
    "Avg RTT DL (ms)",
    "Avg RTT UL (ms)",
    "TCP DL Retrans. Vol (Bytes)",
    "TCP UL Retrans. Vol (Bytes)",
    "DL TP < 50 Kbps (%)",
    "50 Kbps < DL TP < 250 Kbps (%)",
    "250 Kbps < DL TP < 1 Mbps (%)",
    "DL TP > 1 Mbps (%)",
    "UL TP < 10 Kbps (%)",
    "10 Kbps < UL TP < 50 Kbps (%)",
    "50 Kbps < UL TP < 300 Kbps (%)",
    "UL TP > 300 Kbps (%)",
    "HTTP DL (Bytes)",
    "HTTP UL (Bytes)",
    "Nb of sec with 125000B < Vol DL",
    "Nb of sec with 1250B < Vol UL < 6250B",
    "Nb of sec with 31250B < Vol DL < 125000B",
    "Nb of sec with 6250B < Vol DL < 31250B",
    "Nb of sec with Vol DL < 6250B",
    "Nb of sec with Vol UL < 1250B"
]


processed_df = handle_outliers(cleaned_df.copy(), columns)
```
